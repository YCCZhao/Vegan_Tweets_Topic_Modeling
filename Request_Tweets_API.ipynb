```python
config = cnfg.load("/home/ubuntu/Documents/.twitter_config")

oauth = OAuth1(config["consumer_key"],
               config["consumer_secret"],
               config["access_token"],
               config["access_token_secret"])
```

```python
search_url = "https://api.twitter.com/1.1/search/tweets.json"
parameters = {"q": "vegan"}
```

```python
def save_tweets(tweets):
    for tweet in tweets:
        if tweet['retweeted']:
            continue
        
        data = {}
        data['id'] = tweet['id']
        data['datetime'] = tweet['created_at']
        data['text'] = tweet['text']
        data['entities'] = tweet['entities']
        data['reply_to_tweet'] = tweet['in_reply_to_status_id']
        data['user'] = tweet['user']['id']
        data['retweet_count'] = tweet['retweet_count']
        data['favorite_count'] = tweet['favorite_count']
        #data['possibly_sensitive'] = tweet['possibly_sensitive']
        data['lang'] = tweet['lang']
        
        vegan_tweets.insert_one(data)
        
```

```python
response = requests.get(search_url, auth=oauth)
tweets = response.json()['statuses']
```

```python

while 1:
    response = requests.get(search_url, auth=oauth)
    save_tweets(response.json()['statuses'])
    try:
        next_page_url = search_url + response.json()['search_metadata']['next_results']
        response = requests.get(next_page_url, auth=oauth)
    except:
        break
```
